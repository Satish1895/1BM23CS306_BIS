```python
import numpy as np

def gwo(func, lb, ub, dim=1, wolves=10, iters=30):
    lb, ub = np.array(lb), np.array(ub)
    pop = np.random.uniform(lb, ub, (wolves, dim))
    fit = np.array([func(w) for w in pop])
    idx = fit.argsort()
    a, b, d = pop[idx[0]], pop[idx[1]], pop[idx[2]]
    a_score = fit[idx[0]]

    for t in range(iters):
        A = 2 - 2 * t / iters
        for i in range(wolves):
            for j in range(dim):
                r1, r2 = np.random.rand(), np.random.rand()
                X1 = a[j] - (2*r2)*abs(a[j] - pop[i,j])*(2*A*r1 - A)
                r1, r2 = np.random.rand(), np.random.rand()
                X2 = b[j] - (2*r2)*abs(b[j] - pop[i,j])*(2*A*r1 - A)
                r1, r2 = np.random.rand(), np.random.rand()
                X3 = d[j] - (2*r2)*abs(d[j] - pop[i,j])*(2*A*r1 - A)
                pop[i,j] = (X1 + X2 + X3) / 3
            pop[i] = np.clip(pop[i], lb, ub)

        fit = np.array([func(w) for w in pop])
        idx = fit.argsort()
        a, b, d = pop[idx[0]], pop[idx[1]], pop[idx[2]]
        a_score = fit[idx[0]]
        print(f"Iter {t+1}/{iters} | Best Power = {-a_score:.4f}")

    return a, -a_score

def pv_power(x):
    V = x[0]
    Isc = 8
    Voc = 40
    I = Isc * (1 - (V/Voc)**2)
    return -(V * I)

if __name__ == "__main__":
    best_V, best_P = gwo(
        func=pv_power,
        lb=[0],
        ub=[40],
        dim=1,
        wolves=10,
        iters=30
    )
    print("\nBest Voltage:", best_V[0])
    print("Max Power:", best_P)
```
